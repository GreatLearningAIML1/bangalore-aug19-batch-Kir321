2.Preprocess rows of the “text” column

```
## Importing countvectorizer and performing the Lowercase,
## stopwords have been forced. Unwanted spaces and characters are automatically removed unless 
## included using token_pattern 

from sklearn.feature_extraction.text import CountVectorizer
cvect = CountVectorizer(stop_words='english',lowercase=True)
```

```
bl_df.dtypes
```

#### As we want to make this into a multi-label classification now we will add id	gender	age	topic	sign together to form one more table called lable so that we can consider only 2 columns

```
bl_df['labels']=bl_df['gender']+bl_df['age'].astype(str)+bl_df['topic']+bl_df['sign']
bl_df.head(8)
```

## drop id	gender	age	topic	sign and date so that we will have only 2 colums for consideration  

```
bl_df1=bl_df.drop(['id','gender','age','topic','sign','date'],axis=1)
print(bl_df1.head(5))
```

Separating test and lables as x and y so that we can split and train the data 

```
x=bl_df1['text']
y=bl_df1['labels']
print(x.head(5))
print(y.head(5))
```

```
from sklearn.model_selection import train_test_split
```

```
# split X and y into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=10)
```

```
##Training data
print(x_train.shape)
print(y_train.shape)
```

```
#Test data
print(x_test.shape)
print(y_test.shape)
```

5.Feeding input data into CountVectorizer after including ngram using Use ngram_range=(1, 2) 

```
cvect = CountVectorizer(stop_words='english',lowercase=True,ngram_range=(1,2))
cvect.fit(x_train)
```

vocablury Size 

```
len(cvect.vocabulary_)
```

# Convert Training/Test data Count Vectors

```
x_train_ct=cvect.transform(x_train)
x_test_ct=cvect.transform(x_test)
```

7.Multi lable transformation 

```
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
```

```
y_train1=mlb.fit_transform(y_train)
y_train1
```

```
y_test1=mlb.fit_transform(y_test)
y_test1
```

8.using OneVsRestClassifier and LogisticRegression

```
from sklearn.multiclass import OneVsRestClassifier
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(solver='lbfgs',max_iter=1000)
clf=OneVsRestClassifier (clf)
```

```
clf.fit(x_train_ct, y_train1)
```

9.i. Accuracy score 
ii. F1 score

```
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn import preprocessing
from sklearn.metrics import f1_score,classification_report, confusion_matrix,accuracy_score
```

```
y_predict = clf.predict(x_test_ct)
print(y_predict)
```

```
y_predict1=mlb.inverse_transform(y_predict)
print(metrics.accuracy_score(y_test1, y_predict))
```
