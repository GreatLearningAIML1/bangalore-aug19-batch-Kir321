# qustions for ststistical evidence for the given data set 

###  a.Do charges of people who smoke differ significantly from the people who don't

```python
sns.catplot(x='age', y='charges', hue ='smoker', data = df, kind='swarm', height=15, aspect=1)
```

#### From here we can see charges of smokers is high as compared to non-smokers here orenge dots are non smokers which lies in the lowe range of the charges compare to the blue dots which are in higher range of the charges 

### b. Does bmi of males differ significantly from that of females?

```python
sns.catplot(x='bmi', y='age', hue ='sex', data = df, kind='swarm', height=15, aspect=1)
```

### here we can see dots distributed all over the plot so we can say BMI will not differ WRT sex 

### c. Is the proportion of smokers significantly different in different genders?

### for this problem approch is like finding the total number of smokers WRT to gender 

```python
df.groupby(by=['smoker','sex'], as_index=False).count()
```

 ### proportion of smokers does not differ between male and female. Total Male smokers are 159 and total female smokers are 115

### d. Is the distribution of bmi across women with no children, one child and two children, the same ?

```python
print(df[df['children'] == 0].describe())
print(df[df['children'] == 1].describe())
print(df[df['children'] == 2].describe())
```

```python
df.groupby(by=['children','sex'], as_index=False).count()
```

### from here we can see for 1 2 3 children BMI number has changed but not the effect of BMI can be seen for this change
